## Working with Files
Essential techniques for reading from and writing to files in Python, with a special focus on handling geospatial data.
>Always ensure your data files exist before trying to process them. Creating sample data programmatically is also useful for testing your programs and providing examples for others to follow.

```python
sample_data = """35.6895,139.6917
34.0522,-118.2437
51.5074,-0.1278
-33.8688,151.2093
48.8566,2.3522"""

output_file = "coordinates.txt"

try:
    with open(output_file, "w") as file:
        file.write(sample_data)
        print(f"Sample file '{output_file}' has been created successfully.")
except Exception as e:
    print(f"An error occurred while creating the file: {e}")
```

This code example demonstrates several important file handling concepts:
1. **String literals with triple quotes**: The sample_data uses triple quotes (`"""`) string, which is perfect for storing structured data like coordinates.
2. **The `with` statement**: This is Python’s recommended way to work with files. It automatically handles file opening and closing, even if an error occurs. This prevents resource leaks and ensures data is properly saved.
3. **Exception handling**: The `try-except` block catches any errors that might occur during file creation, such as permission issues or disk space problems.

Note that the `write` method is used to write the string to the file. The `w` mode is used to open the file for writing. If the file does not exist, it will be created. If the file exists, it will be overwritten. If you do not want to overwrite the file, you can use the `a` mode to append to the file. <br><br>

Python's file handling capabilities are built into the language. You don't need to import any special libraries for base file operations.

**Key concepts**:
- **Reading files**: Getting data from external files into your program
- **Writing files**: Saving data from your program to external files
- **File modes**: Understanding when to use “read” vs “write” mode
- **Processing data**: Transforming data as you read or write it

```python
# Example of reading coordinates from a file and writing to another file
input_file = "coordinates.txt"
output_file = "output_coordinates.txt"
try:
    # Step 1: Read the coordinate data from our input file
    with open(input_file, "r") as infile:
    # readlines() returns a list where each element is a line from the file
    # Each line still includes the newline character (\n) at the end
        coordinates = infile.readlines()
# Step 2: Process the data and write it to a new file with better formatting
    with open(output_file, "w") as outfile:
        for line in coordinates:
            # strip() removes whitespace (including \n) from both ends of the string
            # split(",") breaks the line into parts wherever it finds a comma
            lat, lon = line.strip().split(",")
            # Write the formatted data to our output file
            # The \n adds a newline character so each coordinate is on its own
            line
            outfile.write(f"Latitude: {lat}, Longitude: {lon}\n")
        print(f"Coordinates have been written to {output_file}")
except FileNotFoundError:
    print(f"Error: The file {input_file} was not found.")
    print("Make sure you ran the previous code cell to create the coordinates.txt file.")

    
```

O método `.read()` vem do `file object` retornado pela função `open()`. Quando `open()` é chamado, ele cria objetos passíveis de leitura com métodos como `.readlines()`, `.readline()` e `.read()`. Todos esses métodos não exibem nada na tela quando chamados. É preciso chamar a função `print()` para exibir o resultado para o usuário. 

`.read()` e `.readline()` retornam objetos da classe `str`, enquando `.readlines()` retorna uma lista em que cada elemento da lista é parte da string que foi lida e contêm o caractere de quebra de linha `\n` como parte do conteúdo.

`.read()` lê todo o conteúdo, `.readline()` lê apenas a primeira linha, `.readlines()` lê todo o conteúdo e armazena cada seção em uma lista.

Um dos comportamentos da função `open()` é criar um cursor de leitura interno. Esse cursor começa na posição 0 e vai até o final do arquivo. Por este motivo, chamar a função print como no bloco de código abaixo, não retorna nada após o primeiro print, pois o cursor já está na última posição. Todo o conteúdo foi consumido.<br><br>
Para reiniciar o cursor, podemos utilizar a função `.seek()` e passar o `int` 0. Dessa forma, o cursor retorna para a posiçao 0. Outra forma seria utilizar apenas um método por abertura e repetir o bloco de código que inicia com `with`.

```python
arquivo = '../data/raw/exercicios_1.txt'
```

```python
# Lendo o arquivo
with open(arquivo, 'r') as file:
    print(file.read())
    print(f'Tipo retornado por .read(): {type(file.read())}')

    file.seek(0)
    print('\n',file.readline())
    print(f'Tipo retornado por .readline(): {type(file.readline())}')

    file.seek(0)
    print('\n',file.readlines())
    print(f'Tipo retornado por .readlines(): {type(file.readlines())}')
```

Quando utilizamos a função `open()` no modo de escrita `w`, funções que envolvem leitura, como `.read()`, não funcionam. Para que seja possível escrever e ler numa única chamada do método `open()`, utilizamos `w+r` como modo de leitura. O conceito de cursor continua válido — após a escrita, é necessário reiniciá-lo para que possamos ler o arquivo.<br><br>Além disso, o file object criado pela função `open()` é indicado logo após o termo `as`, que no exemplo abaixo é chamado de `file`. Os métodos `.read()`, `.write()`, entre outros, são aplicados somente a esse file object (_file_).

```python
texto = 'teste.txt'
saida = 'O texto deste arquivo foi escrito com a função .open() e o método .write()'

with open(texto, 'w') as file:
    file.write(saida) # a var arquivo possui o caminho de um txt na forma de str, por isso o código funciona
```

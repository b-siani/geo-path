### O que é Estatistica? 
É um campo da matemática que reúne métodos para analisar dados e não apenas transformá-los em informação,mas também modela variabilidade, quantifica incerteza e permite inferência sobvre populações. Uma estatística por sua vez, é uma medida numérica calculada a partir de um conjunto de dados (amostra), como média ou desvio padrão, utilizada para resumir ou inferir características sobre os dados.<br><br>No campo da estatística, é comum calcular medidas que sintetizam informações de uma amostra e fornecem mais ocntexto sobre o fenômeno obsevado. Uma estatística, por sua vez, é uma medida numérica calculada a partir de um conjunto de dados, como média, variância ou desvio padrão, utilizada para resumir ou inferir características desses dados.<br><br>De forma didática, pode-se pensar na estatística como lidando com dois aspectos principais: a análise da variabilidade e a quantificação da incerteza. Ao estmimar característica de uma população, é fundamental mensurar o grau de incerteza para que saibamos o quão confiável é a estimativa em relação a medida real. Isso é de extrema importância para que possamos tomar decisões com o maior nível de confiança baseado nos dados.<br><br>Além disso, a análise da variância permite compreender como os valores estão distribuídos em torno de um valor central. Através da medida de variação da amostra, obtemos mais contexto sobre o objeto de estudo e sobre o valor central, por exemplo: "70% dos Brasileiros estão endividados", "35% tem entre 25 e 35 anos", essa segunda informação nos dá contexto e permite inferir que pouco mais de um terço da população endividada é composta por jovens adultos entre 25 e 35 anos, restando ~2/3 da população acima dos 35 com endividamentos (aqui os 35% são condicionados ao grupo dos endividads).<br><br>Por fim, a estatística trata dos métodos de coleta de dados. Técnicas de amostragem permitem economizar recursos ao definir o tamanho mínimo de uma amostra e os procedimentos necessários para para garantir representativade e reduzir vieses, otimizando recursos e aumentando a confiabilidade das conclusões.

### De onde vem os dados? 
Existem duas formas principais:
1. **Process / organic data**, dados gerados de forma orgânica ao longo do tempo como resultado de um processo. Exemplo: dados de navegação na internet, transações bancárias, performance de indivíduos em algum esporte, histórico de visualização no Netflix, dados de temperatura. O termo **Big Data** está atrelado a dados provenientes de processos orgânicos, os quais geram conjuntos de dados massivos. 
2. "Designed" data, dados coletados a partir de uma amostra a qual é objeto de pesquisa. Os indivíduos são amostras da população alvo. Os conjuntos de dados são tipicamente menores, e as informações são coletadas por motívos específicos. 


Ao analisar dados, independente da fonte, é importante se perguntar, 
> Os dados podem ser considerados i.i.d.?<br>
> i = indepentende e id = identicamente distribuído

Observações da variável de interesse (quando é i.i.d.) são completamente independentes de todas observações. Não há correlação entre as diferentes medidas que estão sendo coletadas com as das unidades da populaçõa. Todas as observações são indepentendentes (operações diferentes na bolsa, jogos diferentes, indivíduos diferentes numa pesquisa). Além disso, todos os valores observados podem ser descritos por uma distribuição estatística comum. 

**Variável Quantitativa Discreta**

- Valores numéricos inteiros (não admitem frações intermediárias).
  - Número de árvores em um lote: 12, 25, 37
  - Número de filhos por família: 0, 1, 2, 3
  - Quantidade de parcelas amostradas: 10, 15, 20

**Variável Quantitativa Contínua**

- Valores numéricos que podem assumir qualquer valor em um intervalo (incluindo decimais).
  - Altura de árvores (m): 12,3; 15,7; 18,25
  - Diâmetro à altura do peito (DAP) (cm): 10,5; 22,8; 35,1
  - Temperatura (°C): 18,2; 21,6; 30,0

**Variável Qualitativa Nominal**

- Categorias sem ordem natural.
  - Espécie de árvore: Ipê, Eucalipto, Pinus
  - Cor da folha: verde, amarela, marrom
  - Tipo de solo: arenoso, argiloso, siltoso

**Variável Qualitativa Ordinal**

- Categorias com ordem definida, mas sem **escala** numérica exata. 
- Algo que pode ajudar a distinguir casos que induzem ao erro — por exemplo, presença de números para representar rótulos — é pensar na distância entre os números; entre 1 e 2, 5 e 6, a distância é de uma unidade, porém, entre satisfeito e extremamente satisfeito não é diretamente mensurável.
  - Estágio de crescimento: muda, jovem, adulta
  - Nível de infestação: baixo, médio, alto
  - Classificação de qualidade da madeira: ruim, regular, boa, excelente

### Tipos de estudos
- **Confirmatory vs Exploratory Research**
<br><br>
**Confirmatory**: Há uma hipóteses que queremos testar e toda a análise de dados gira em torno deste teste para validar se nosssa hipótese é verdadeira ou falsa.<br>
**Exploratory**: Não há uma pergunta a ser respondida, a análise de dados é feita sem uma hipótese em mente.
"Quanto maior o número de perguntas feitas para um dataset, maiores são as chances de chegar a uma conclusão errônea"
<br><br>
Frequentemente, o objetivo dos estudos é comparar e contrastar uma quantidade em relação a outra. Por exemplo, produção de madeira em um sítio com e sem adubação, preferência de votos entre candidados, taxa de cliques entre duas versões de um mesmo design, etc.
<br><br>
- **Non-Comparative Research Studies**
<br><br>
O foco desse tipo de estudo é estimar ou predizer quantidades absolutas (não é explicitamente comparativo). Ex., aumento ou queda de ações daqui um ano, pressão arterial sob uso de um determinado remédio.
<br><br>
- **Observational Data and Experiments**
<br><br>
Os dados de **estudos observacionais** são gerados naturalmente. A seleção e constraste de indivíduos ocorre naturalmente dentro dos grupos. Podemos comparar notas entre estudantes em salas com professores com mais ou menos experiência, _life span_ enre fumantes e não fumantes, etc. Nesse tipo de estudo, é comum dizer que os indivíduos foram "expostos" a uma certa condição ao invés de serem designados.<br><br>
Já em **experimentos**, há manipulação ou designação, de tal forma que o próprio experimento trata unidades de formas diferentes (ex., da produçào de madeira, o sítio é deliberadamente diferente). É comum que experimentos estejam atrelados a testes aleatórios e, por consequência, os indivíduos sejam designados aleatóriamente a uma condição.

#### Quantitative data: Histograms
Histogramas sao comumente utilizados para uma primeira visualização dos dados quantitativos, eles são excelentes para entender como nossos dados se parecem e como proceder com as análises.<br>Histogramas possuem 4 aspectos principais:
- Forma: aparência geral, permite observar se os dados seguem uma distribuição normal (em forma de sino), simétrica, ou assimétrica (à direita ou à esquerda, de acordo com a cauda). 
  - Quando a distribuição é assimétrica, é interessante dizer quanto da distribuição da variável observada está numa determinada região
  - Quando a distribuição possui apenas um pico, chamamos de unimodal.<br><br>

- Centro: média ou mediana
- Dispersão: quão distribuídos estão os dados, para isso observamos a amplitude, amplitude interquartis (IQR - Interquartile Range), desvio padrão e variância.
- Outliers: dados dispostos nos extremos da massa de dados.<br>
**OBS: Gráficos de barras != Histogramas.**

#### Pontuação padrão (z-score) e a "Regra Empírica"
A pontuação padrão, é uma medida estatística que indica a quantos desvios padrão ($\sigma$) um determinado valor observado ($x$) está em relação a média ($\mu$) de um conjunto de dados.<br>

$$\large
Z = \frac{x - \mu}{\sigma}
$$

Quando analisamos uma distribuição normal (formato de sino), conseguimos observar regiões que agrupam parte dos nosso dados de acordo com relações entre a dispersão dos dados em torno da média com base no desvio padrão. Por exemplo, imaginemos que a quantidade de sono das pessoas siga uma distribuição normal, que a média de sono seja igual a 7 e o desvio padrão igual a 1,7.<br>
Quando nos afastamos **um desvio padrão em relação a média**, encontramos os valores 5,3 e 8,7. Se, nestes pontos, traçarmos uma linha reta no eixo Y, criaremos uma região na distribuição normal na qual a área sob a curva a um desvio padrão da média representa **68%** de todas as observações. Quando repetimos esse processo e observamos os valores a **dois desvios padrão**, temos uma fatia que representa **95%** dos dados. Por fim, ao observar os dados compreendidos a **três desvios padrão** da média, obtemos um conjunto composto por **99.7%** de todo o conjunto de dados.<br><br>
Essa relação recebe o nome de "Regra Empírica" ou "68-95-99.7". Para uma distribuição normal, quase todos os dados estão distribuídos em uma amplitude de até 3 desvios padrão em relação a média. 

A pontuação padrão (ou z-score), nos diz o quão comum ou raro é um determinado resultado em relação a média. O valor de $z$ permite padronizar dados para comparação, onde z = 0 é a média, z > 0 é acima da média, e z < 0 é abaixo da média. Outra forma de interpretar o valor Z, é como desvio padrão em relação a média, por exemplo, um valor de Z = 1,25 indica que o valor está a 1,25 desvios padrão em relação a média, se fosse negativo, seria o mesmo número absoluto, porém, abaixo da média.<br><br>
Exemplo: costumo dormir 8 horas de sono. Com base nos valores acima meu z-score seria calculado da seguinte forma:
$$Z = \frac{8 - 7}{1.7} = 0.59$$
Portanto, para 8 horas de sono, z-score = 0.59 é um valor relativamente comum, está a menos de 1 desvio padrão e é um valor positivo em relação a média.<br><br>Se ao invés do número de horas de sono, fosse dado um valor qualquer de desvio padrão (-2,7) e tivéssemos que obter o número de horas que corresponde aquele valor, o cálculo seguiria da seguinte forma:
$$-2.7 = \frac{x - 7}{1.7}$$
$$-2.7 \times 1.7 = x - 7$$
$$x = 2.41$$



### Quantitative Data Graphical Summary: BoxPlot
#### What is a Boxplot?
Um Boxplot é uma representação gráfica que sumariza nossos dados. O Boxplot é composto pelas métricas de mínimo, Q1, mediana, Q3, máximo.<br><br> 

Uma das vantagens de se utilizar um boxplot, é visualizar outliers, pois sua representação é exibida separadamente. A visualização de "conjuntos" de dados também é melhor com boxplots. Entretanto, boxplot sozinho não é suficiente para representação gráfica dos dados pois pode ocultar "gaps" e grupos nos dados observados.
